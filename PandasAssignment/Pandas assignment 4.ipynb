```python
states = pd.read_csv(r"C:\Datasets\Nigeria data\list-states-nigeria-227j.csv")
states
```

```python
poverty_rate = pd.read_csv(r"C:\Datasets\Nigeria data\Nigeria povertyrate per local gov.csv")
poverty_rate
```

```python
clean_pr = poverty_rate.drop(['ID_0', 'ISO', 'NAME_0', 'ID_12'], axis=1)
clean_pr
```

```python
clean_pr.info()
```

```python
combined = pd.merge(clean_pr, local_gov, left_on = 'NAME_23', right_on = 'Local Government Area', how='inner')
combined
```

```python
combined.info()
```

```python
clean_combined = combined.drop(['NAME_12', 'NAME_23'], axis=1)
clean_combined
```

```python
clean_states = states.dropna(subset=['Rank'], axis=0)
clean_states = clean_states.drop(['SNo', 'Rank', 'Unnamed: 4'], axis=1)
clean_states
```

```python
clean_states['State'] = clean_states['State'].replace('Abuja Federal Capital Territory', 'FCT')
clean_states
```

```python
merged_data = pd.merge(clean_combined, clean_states, on= 'State', how='left')
merged_data
```

```python
merged_data.info()
```

```python
null_val = merged_data[merged_data['Population'].isna()]
null_val
```

```python
merged_data['Population'] = pd.to_numeric(merged_data['Population'].str.replace(',', ''), errors='coerce')
```

```python
# EFFECT OF POPULATION ON POVERTY RATE BY LOCAL GOVERNMENT 
grouped_data = merged_data.groupby('Local Government Area')
result_groupby = grouped_data[['Population', 'Poverty_Ra']].mean()
result_groupby
```

```python
result_groupby = result_groupby.sort_values(by='Poverty_Ra')
result_groupby
```
