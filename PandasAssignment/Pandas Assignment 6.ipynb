QUESTION 
Using the Melbourne_housing_FULL.csv

1. Extract from the dataset where Price column is not null into another variable

2. Based on the this new variable clean the data to ensure no more NAN on the dataset by applying the following:
    * Delete unuseful columns from the dataset
    * Use the number of Rooms column to fill for mising number of Bedroom2
    * Use the number of bedroom2 column to fill for the mising number of Bathroom
    * Using this same logic, look for a column that have close relationship with another column with missing number and           use it to fill for the missing values in the column

3. Show in a table, how Bedroom2 and landsize affect the price of house

4. Also show how BuildingArea and YearBuilt affect the price of house

5. In a tabel, show the sum and media of price of house based on Bedroom2

```python
import pandas as pd
import numpy as np

melbourne_housing = pd.read_csv(r"C:\Datasets\Melbourne_housing_FULL.csv")
melbourne_housing
```

```python
non_null = melbourne_housing[melbourne_housing['Price'].notna()]
non_null
```

```python
non_null.info()
```

```python
# Use the number of Rooms column to fill for mising number of Bedroom2
non_null.loc[:, 'Bedroom2'] = non_null['Bedroom2'].fillna(non_null['Rooms'])
non_null.Bedroom2
```

```python
non_null.info()
```

```python
# Use the number of bedroom2 column to fill for the mising number of Bathroom
non_null.loc[:, 'Bathroom'] = non_null['Bathroom'].fillna(non_null['Bedroom2'])
non_null.Bathroom
```

```python
non_null.info()
```

```python
non_null.loc[:, 'BuildingArea'] = non_null['BuildingArea'].fillna(non_null['Landsize'])
```

```python
non_null.loc[:, 'Landsize'] = non_null['Landsize'].fillna(non_null['BuildingArea'])
```

```python
non_null.info()
```

```python
non_null = non_null.dropna(subset=['YearBuilt'])
```

```python
non_null.describe()
```

```python
non_null.loc[:, 'YearBuilt'] = 10 * (non_null['YearBuilt'] // 100)
```

```python
non_null.loc[:, 'YearBuilt'] = non_null['YearBuilt'] / 10
```

```python
non_null['YearBuilt'].unique()
```

```python
# Show in a table, how Bedroom2 and landsize affect the price of house
bedroom = pd.cut(non_null['Bedroom2'], [0, 3, 6, 9, 12])
land_size = pd.cut(non_null['Landsize'], [0, 10000, 20000, 30000, 40000, 50000])
non_null.pivot_table('Price', [bedroom], land_size, aggfunc= {'Price': sum})
```

```python
non_null.pivot_table('Price', land_size, bedroom, aggfunc= {'Price': sum})
```

```python
non_null.groupby([bedroom, land_size])['Price'].sum().unstack()
```

```python
non_null.groupby([land_size, bedroom])['Price'].sum().unstack()
```

```python
building_area = pd.cut(non_null['BuildingArea'], [0, 10000, 20000, 30000])
# year_built =
non_null.pivot_table('Price', 'YearBuilt', building_area, aggfunc = {'Price': sum})
```
