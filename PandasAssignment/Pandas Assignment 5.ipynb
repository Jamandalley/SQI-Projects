```python
import pandas as pd
import numpy as np

california_city = pd.read_csv(r"C:\Datasets\california_cities.csv")
california_city
```

```python
california_city = california_city.drop(labels=['Unnamed: 0', 'Unnamed: 14', 'Unnamed: 15'], axis=1)
california_city
```

```python
california_city = california_city.set_index('city')
california_city
```

```python
california_city.describe()
```

```python
# Extract the city with highest population total, highest area land square meter and area water square meter
highest_population_city = california_city['population_total'].idxmax()
highest_land_area = california_city['area_land_sq_mi'].idxmax()
highest_water = california_city['area_water_sq_mi'].idxmax()
print('The city with the highest population is: ', highest_population_city)
print('The city with the highest area land square meter is: ', highest_land_area)
print('The city with the highest area water square meter is: ', highest_water)
```

```python
# Calculate the total area total square meter of California city.
total_area = california_city['area_total_sq_mi'].sum()
print('The Total Area Square Meter of California city is: ', total_area)
```

```python
# Calculate the sum of the dataset along column axis
variables_sum = california_city.select_dtypes(include='number').sum()
variables_sum
```
