# DATA CLEANING, PREPARATION AND ANALYSIS
1. UNEMPLOYMENT RATE

```python
unemployment_rate.describe()
```

```python
# Since we're only interested in the unemployment rate, use the unemployment rate column to filter the table
unemployment_rate = unemployment_rate.dropna(subset=['Unemployment, female (% of female labor force) (modeled ILO estimate)', 'Unemployment, male (% of male labor force) (modeled ILO estimate)'])
Africa_unemployment_rate = pd.merge(unemployment_rate, country_code, left_on='Entity', right_on='name', how='left')
columns_to_drop = ['name', 'alpha-2', 'alpha-3', 'country-code', 'sub-region', 'Continent']
columns_to_drop = [col for col in columns_to_drop if col in Africa_unemployment_rate.columns]
Africa_unemployment_rate.drop(columns=columns_to_drop, inplace=True)
Africa_unemployment_rate = Africa_unemployment_rate.rename(columns={'region': 'Continent', 'Entity': 'Country'})
Africa_unemployment_rate.reset_index()
```

```python
Africa_unemployment_rate['Continent'] = Africa_unemployment_rate['Continent'].str.strip()
A_unemployment_rate = Africa_unemployment_rate[Africa_unemployment_rate['Continent'] == 'Africa']
A_unemployment_rate
```

```python
A_unemployment_rate.info()
```

```python
A_unemployment_rate.isnull().sum()
```

```python
africa_unemployment_rate = A_unemployment_rate.copy()
africa_unemployment_rate.rename(columns={'Unemployment, female (% of female labor force) (modeled ILO estimate)': 'Unemployment in female',
                                        'Unemployment, male (% of male labor force) (modeled ILO estimate)': 'Unemployment in male',
                                         'Population (historical estimates)': 'Population'}, inplace=True)
africa_unemployment_rate.drop(columns='Continent')
africa_unemployment_rate
```

```python
# PREVALENCE OF UNEMPLOYMENT IN AFRICA
prevalence_of_unemployment = africa_unemployment_rate.groupby(['Country'])[['Unemployment in female', 'Unemployment in male']].mean()
prevalence_of_unemployment['Average Prevalence'] = (prevalence_of_unemployment['Unemployment in female'] + prevalence_of_unemployment['Unemployment in male'])/2
prevalence_of_unemployment = prevalence_of_unemployment.sort_values(by='Average Prevalence', ascending=False)
prevalence_of_unemployment.head()
```

```python
import warnings
warnings.filterwarnings("ignore")
df = prevalence_of_unemployment.head()
df['country'] = np.array(df.index)
df
```

```python
df1 = prevalence_of_unemployment.tail().sort_values(by='Average Prevalence')
df1['country'] = df1.index.values
df1
```

```python
# Bar Chart Representation of Top 5 countries with the highest prevalence of Unemployment 
%matplotlib inline
sns.set(style='ticks', color_codes=True)
sns.set(style='darkgrid')
fig, ax = plt.subplots(1, 2, figsize =(13, 4))
#For non-binarized data
sns.barplot(x='country', y='Average Prevalence', data=df, ax=ax[0])
ax[0].set_title('Top 5 African Countries with the highest Unemployment Rate')
ax[0].set_xlabel('Countries')
ax[0].set_ylabel('Average Prevalence')
for p in ax[0].patches:
    ax[0].annotate(format(p.get_height(), '.2f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 10), 
                   textcoords = 'offset points')



sns.barplot(x='country', y='Average Prevalence', data=df1, ax=ax[1])
ax[1].set_title('Top 5 African Countries with the Least Unemployment Rate')
ax[1].set_xlabel('Countries')
ax[1].set_label('Average Prevalence')
for p in ax[1].patches:
    ax[1].annotate(format(p.get_height(), '.2f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 10), 
                   textcoords = 'offset points')


plt.show()
```

OBSERVATIONS AND RECOMMENDATIONS
The current analysis of unemployment data in Africa has uncovered notable trends, particularly highlighting a surge in unemployment prevalence within the working-age population as the population increases, notably in the years 2020 and 2021. This surge is predominantly attributed to Djibouti and South Africa. Additionally, the study sheds light on the challenges faced by the majority of African countries in implementing national strategies for youth employment. A concerning observation is the rapid decrease in the number of countries successfully operationalizing such strategies. The analysis also reveals a sharp decline in educational budgets, particularly during 2020 and 2021 when the highest unemployment rates were recorded. Furthermore, a disproportionate increase in the creation of private limited liability companies was observed. 

A. Key Observations:

1. Unemployment Prevalence:

There is a discernible upswing in unemployment prevalence, particularly prominent in 2020 and 2021.
Djibouti and South Africa significantly contribute to this surge. A similar trend in unemployment rate was alos reproted by Statista (2023). 

2. Youth Employment Strategies:

Most African countries face challenges in implementing national strategies for youth employment.
A rapid decline is observed in the successful operationalization of these strategies.

3. Educational Budgets:

A sharp decrease in educational budgets is noted, coinciding with the peak of unemployment rates in 2020 and 2021.

4. Private Limited Liability Companies:

There is a disproportionate increase in the establishment of private limited liability companies.

These observations are corroborated by findings from a systematic review conducted by Ongbali, Afolalu and UdO (2019) whereby unemployemnt rate was found to be uderpinned byfast population growth, rural to urban migration and lack of employable skills and experience among the youth fast population growth, rural to urban migration and lack of employable skills and experience among the youth.

B. RECOMMENDATIONS: 

1. Operationalization of Youth Employment Strategies:

Urgent attention and action are needed to operationalize youth employment and empowerment strategies across African nations.

2. Increase in Educational Budget:

A strategic increase in educational budgets is recommended to address the challenges posed by the observed downturn.

C. LIMITATIONS:
1. Data Disproportion:

The datasets exhibit a disproportion in distribution, with not all countries uniformly represented across all study years.

2. Correspondence Disparities:

Challenges in value representation hindered the optimum correspondence of incidence.

D. CONCLUSION:
This analysis provides valuable insights into the unemployment landscape in Africa. Despite the limitations identified, the findings lay the groundwork for informed decision-making. To enhance the richness of our analysis, it is strongly recommended that incorporating datasets from other countries or continents with lower incidence rates for a comprehensive comparison of incidence patterns and underlying factors should be considered.
